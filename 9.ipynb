```python
1. Given a 1-indexed array of integers numbers that are already sorted in non-decreasing order,
find two numbers such that they add up to a specific target number. Let these two numbers be
numbers[index1] and numbers[index2] where 1 <= index1 < index2 < numbers.length.
Return the indices of the two numbers, index1, and index2, added by one as an integer
array [index1, index2] of length 2.
The tests are generated such that there is exactly one solution. You may not use the
same element twice.
Your solution must use only constant extra space.
Input: numbers = [2,7,11,15], target = 9
Output: [1,2]
Explanation: The sum of 2 and 7 is 9. Therefore, index1 = 1, index2 = 2. We return [1, 2].

Ans. def two_sum(numbers, target):
    left, right = 0, len(numbers) - 1

    while left < right:
        current_sum = numbers[left] + numbers[right]

        if current_sum == target:
            return [left + 1, right + 1]  # Adding 1 to convert from 0-indexed to 1-indexed
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    # If no solution is found, the function would not reach the return statement above.
    # In such cases, you can either raise an exception or return None to handle the situation.
    return None

# Example usage:
numbers = [2, 7, 11, 15]
target = 9
result = two_sum(numbers, target)
print(result)  # Output: [1, 2]

2. Given an array of integer nums sorted in non-decreasing order, find the starting and
ending position of a given target value.
If the target is not found in the array, return [-1, -1].
You must write an algorithm with O(log n) runtime complexity
Input: nums = [5,7,7,8,8,10], target = 8
Output: [3,4]

Ans. def find_target_range(nums, target):
    def find_left_boundary():
        left, right = 0, len(nums) - 1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return left

    def find_right_boundary():
        left, right = 0, len(nums) - 1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] <= target:
                left = mid + 1
            else:
                right = mid - 1
        return right

    left_boundary = find_left_boundary()
    right_boundary = find_right_boundary()

    if left_boundary <= right_boundary:
        return [left_boundary, right_boundary]
    else:
        return [-1, -1]

# Example usage:
nums = [5, 7, 7, 8, 8, 10]
target = 8
result = find_target_range(nums, target)
print(result)  # Output: [3, 4]

3. A peak element is an element that is strictly greater than its neighbors.
Given a 0-indexed integer array nums, find a peak element, and return its index. If the array
contains multiple peaks, return the index to any of the peaks.
You may imagine that nums[-1] = nums[n] = -âˆž. In other words, an element is always considered
to be strictly greater than a neighbor that is outside the array.
You must write an algorithm that runs in O(log n) time.
Input: nums = [1,2,3,1]
Output: 2
Explanation: 3 is a peak element and your function should return the index number 2.

Ans. def find_peak_element(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[mid + 1]:
            # If the middle element is greater than the next element, there might be a peak on the left side.
            right = mid
        else:
            # Otherwise, there might be a peak on the right side, or the middle element is part of a decreasing sequence.
            left = mid + 1

    # At the end of the loop, left and right will converge to the index of a peak element.
    return left

# Example usage:
nums = [1, 2, 3, 1]
result = find_peak_element(nums)
print(result)  # Output: 2

4. Given a sorted array of distinct integers and a target value, return the index if the
target is found. If not, return the index where it would be if it were inserted in order.
You must write an algorithm with O(log n) runtime complexity.
Input: nums = [1,3,5,6], target = 5
Output: 2

Input: nums = [1,3,5,6], target = 7
Output: 4

Ans. def search_insert_position(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    # If the target is not found, the left pointer will indicate the index where it should be inserted.
    return left

# Example usage:
nums = [1, 3, 5, 6]
target = 5
result = search_insert_position(nums, target)
print(result)  # Output: 2

target = 7
result = search_insert_position(nums, target)
print(result)  # Output: 4

5. Find the majority element in the array. A majority element in an array A[] of size n is
an element that appears more than n/2 times (and hence there is at most one such element).
Input: A[]={3, 3, 4, 2, 4, 4, 2, 4, 4}
Output: 4
Explanation: The frequency of 4 is 5 which is greater than half of the size of the array size.

Ans. def find_majority_element(nums):
    # Step 1: Find the potential candidate for the majority element using Boyer-Moore's voting algorithm.
    candidate = None
    count = 0

    for num in nums:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1

    # Step 2: Verify if the candidate is the actual majority element by counting its occurrences.
    count = nums.count(candidate)
    if count > len(nums) // 2:
        return candidate
    else:
        return None  # No majority element found.

# Example usage:
nums = [3, 3, 4, 2, 4, 4, 2, 4, 4]
result = find_majority_element(nums)
print(result)  # Output: 4

6. You are a product manager and currently leading a team to develop a new product. Unfortunately,
the latest version of your product fails the quality check. Since each version is developed based
 on the previous version, all the versions after a bad version are also bad.
Suppose you have n versions [1, 2, ..., n] and you want to find out the first bad one, which causes
all the following ones to be bad.
You are given an API bool isBadVersion(version) which returns whether the version is bad. Implement
a function to find the first bad version. You should minimize the number of calls to the API.
Input: n = 5, bad = 4
Output: 4
Explanation:
call isBadVersion(3) -> false
call isBadVersion(5) -> true
call isBadVersion(4) -> true
Then 4 is the first bad version.

Ans. def first_bad_version(n):
    left, right = 1, n

    while left < right:
        mid = left + (right - left) // 2

        if isBadVersion(mid):
            # If the middle version is bad, the first bad version must be on or before this version.
            right = mid
        else:
            # If the middle version is good, the first bad version must be after this version.
            left = mid + 1

    # At the end of the loop, left and right will converge to the first bad version.
    return left

# Example usage:
n = 5
bad = 4

# Assume you have an isBadVersion API provided.
def isBadVersion(version):
    return version >= bad

result = first_bad_version(n)
print(result)  # Output: 4

7. Given an array of integers, find the inversion of an array. Formally, two elements a[i]
and a[j] form an inversion if a[i] > a[j] and i < j.
N=5, arr[] = {2, 4, 1, 3, 5}
Output: 3
Explanation: (2,1) (4,1) and (4,3) forms an inversion in an array

Ans. def merge(arr, temp, left, mid, right):
    inversions = 0
    i = left
    j = mid + 1
    k = left

    while i <= mid and j <= right:
        if arr[i] <= arr[j]:
            temp[k] = arr[i]
            i += 1
        else:
            temp[k] = arr[j]
            inversions += (mid - i + 1)  # Counting the inversions
            j += 1
        k += 1

    while i <= mid:
        temp[k] = arr[i]
        i += 1
        k += 1

    while j <= right:
        temp[k] = arr[j]
        j += 1
        k += 1

    for i in range(left, right + 1):
        arr[i] = temp[i]

    return inversions

def merge_sort(arr, temp, left, right):
    inversions = 0
    if left < right:
        mid = (left + right) // 2

        inversions += merge_sort(arr, temp, left, mid)
        inversions += merge_sort(arr, temp, mid + 1, right)
        inversions += merge(arr, temp, left, mid, right)

    return inversions

def count_inversions(arr):
    n = len(arr)
    temp = [0] * n
    return merge_sort(arr, temp, 0, n - 1)

# Example usage:
arr = [2, 4, 1, 3, 5]
result = count_inversions(arr)
print(result)  # Output: 3

8. Given three arrays sorted in non-decreasing order, print all common elements in these arrays.
ar1[] = {1, 5, 10, 20, 40, 80}
ar2[] = {6, 7, 20, 80, 100}
ar3[] = {3, 4, 15, 20, 30, 70, 80, 120}
Output: 20, 80

Ans. def find_common_elements(ar1, ar2, ar3):
    common_elements = []
    i, j, k = 0, 0, 0

    while i < len(ar1) and j < len(ar2) and k < len(ar3):
        if ar1[i] == ar2[j] == ar3[k]:
            common_elements.append(ar1[i])
            i += 1
            j += 1
            k += 1
        elif ar1[i] < ar2[j]:
            i += 1
        elif ar2[j] < ar3[k]:
            j += 1
        else:
            k += 1

    return common_elements

# Example usage:
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
result = find_common_elements(ar1, ar2, ar3)
print(result)  # Output: [20, 80]
```
