```python
1.Make the String Great
Ans. def makeGreat(s):
    stack = []
    for char in s:
        if stack and char == stack[-1]:
            stack.pop()
        else:
            stack.append(char)
    return "".join(stack)

2. Remove all adjacent duplicates in a String
Ans. def removeDuplicates(s):
    stack = []
    for char in s:
        if stack and char == stack[-1]:
            stack.pop()
        else:
            stack.append(char)
    return "".join(stack)

3. Online Stock Span
Ans. class StockSpanner:
    def __init__(self):
        self.prices = []
        self.spans = []

    def next(self, price):
        span = 1
        while self.prices and price >= self.prices[-1]:
            span += self.spans.pop()
            self.prices.pop()
        self.prices.append(price)
        self.spans.append(span)
        return span

4. Time Needed to buy tickets
Ans. def mincostTickets(days, costs):
    n = len(days)
    durations = [1, 7, 30]
    dp = [float('inf')] * (n + 1)
    dp[n] = 0

    for i in range(n - 1, -1, -1):
        for j in range(len(durations)):
            k = i
            while k < n and days[k] < days[i] + durations[j]:
                k += 1
            dp[i] = min(dp[i], dp[k] + costs[j])

    return dp[0]

5. Product of the last K Numbers
Ans. class ProductOfNumbers:
    def __init__(self):
        self.prefix_product = [1]

    def add(self, num):
        if num == 0:
            self.prefix_product = [1]
        else:
            self.prefix_product.append(self.prefix_product[-1] * num)

    def getProduct(self, k):
        if k >= len(self.prefix_product):
            return 0
        return self.prefix_product[-1] // self.prefix_product[-k-1]

6. Largest Rectangle in Histogram
Ans. def largestRectangleArea(heights):
    stack = []
    max_area = 0
    i = 0

    while i < len(heights):
        if not stack or heights[i] >= heights[stack[-1]]:
            stack.append(i)
            i += 1
        else:
            top = stack.pop()
            width = i if not stack else i - stack[-1] - 1
            max_area = max(max_area, heights[top] * width)

    while stack:
        top = stack.pop()
        width = i if not stack else len(heights) - stack[-1] - 1
        max_area = max(max_area, heights[top] * width)

    return max_area

7. Sliding Window Maximum
Ans. from collections import deque

def maxSlidingWindow(nums, k):
    result = []
    window = deque()

    for i, num in enumerate(nums):
        while window and nums[window[-1]] < num:
            window.pop()

        window.append(i)

        if i - window[0] >= k:
            window.popleft()

        if i >= k - 1:
            result.append(nums[window[0]])

    return result

8. Design a circular queue data structure with the following operations: enQueue,
deQueue, Front, Rear, and isEmpty.

Ans. class MyCircularQueue:
    def __init__(self, k):
        self.queue = [None] * k
        self.front = 0
        self.rear = -1
        self.size = 0
        self.capacity = k

    def enQueue(self, value):
        if not self.isFull():
            self.rear = (self.rear + 1) % self.capacity
            self.queue[self.rear] = value
            self.size += 1
            return True
        return False

    def deQueue(self):
        if not self.isEmpty():
            self.queue[self.front] = None
            self.front = (self.front + 1) % self.capacity
            self.size -= 1
            return True
        return False

    def Front(self):
        return self.queue[self.front] if not self.isEmpty() else -1

    def Rear(self):
        return self.queue[self.rear] if not self.isEmpty() else -1

    def isEmpty(self):
        return self.size == 0

    def isFull(self):
        return self.size == self.capacity
```
